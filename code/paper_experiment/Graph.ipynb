```python
import matplotlib.pyplot as plt
from scipy import io
import pylab as pb
import pandas as pd
import numpy as np
```

```python
mat_file1 = io.loadmat('2014_Dataset1.mat')
t = mat_file1['t']
p1 = mat_file1['p1']
p2 = mat_file1['p2']
```

```python
for i in range(len(p2)):
    if p2[i]<190:
        print(i)
```

```python
fig = pb.figure(figsize=(14,12))
fig.add_subplot(111, frameon=False)
plt.tick_params(labelcolor='none', top='off', bottom='off', left='off', right='off')
plt.grid(False)



ax1 = fig.add_subplot(211)
ax1.plot(t[0:288]*24,p1[288:576],lw=2)
ax1.set_xlim([0,24])
ax1.set_ylabel(r'$P_{1}$',fontsize=24)
ax1.tick_params(labelsize=20)

ax2 = fig.add_subplot(212)
ax2.set_xlabel(r'$Time$',fontsize=22)
ax2.plot(t[0:288]*24,p2[288:576],lw=2)
ax2.set_xlim([0,24])
ax2.set_ylabel(r'$P_{2}$',fontsize=24)
ax2.tick_params(labelsize=20)

fig.savefig('1.png',bbox_inches='tight')
plt.show()
```
