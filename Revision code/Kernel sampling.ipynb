```python
import GPy
import numpy as np
import matplotlib.pyplot as plt
import pylab as pb
import seaborn as sns
import matplotlib
# matplotlib.rcParams['text.usetex'] = True
sns.set()
```

```python
x = np.array([[0.]])
y = np.array([[0.]])
```

```python
K1 = GPy.kern.Matern52(input_dim=1, lengthscale=0.2)
K2 = GPy.kern.RBF(input_dim=1, lengthscale=0.4)

model1 = GPy.models.GPRegression(x, y, kernel= K1)
model2 = GPy.models.GPRegression(x, y, kernel= K2)

sample_x = np.linspace(0, 3, 100)[:,None]

sample1 = model1.posterior_samples_f(sample_x, size = 2)
sample2 = model2.posterior_samples_f(sample_x, size = 2)

u11 = sample1[:,:,0]
u21 = sample1[:,:,1]
u12 = sample2[:,:,0]
u22 = sample2[:,:,1]
```

```python
plt.figure(figsize=(8,6))
plt.subplot(4,1,1)
plt.plot(sample_x, u11.flatten())
plt.subplot(4,1,2)
plt.plot(sample_x, u21.flatten())
plt.subplot(4,1,3)
plt.plot(sample_x, u12.flatten())
plt.subplot(4,1,4)
plt.plot(sample_x, u22.flatten())
```

```python
a111 = 0.1
a211 = 0.9
a121 = 0.9
a221 = 0.1
a112 = 0.9
a212 = 0.1
a122 = 0.1
a222 = 0.9
```

```python
# q = 1, r = 1
q1r1f1 = a111*u11 
q1r1f2 = a121*u11

# q = 1, r = 2
q1r2f1 = a111*u11 + a211*u21 
q1r2f2 = a121*u11 + a221*u21

# q = 2, r = 1
q2r1f1 = a111*u11 + a112*u12
q2r1f2 = a121*u11 + a122*u12

# q = 2, r = 2
q2r2f1 = a111*u11 + a211*u21 + a112*u12 + a212*u22
q2r2f2 = a121*u11 + a221*u21 + a122*u12 + a222*u22
```

```python
fig1 = pb.figure(figsize=(8,6))
ax1 = fig1.add_subplot(211)
ax1.set_xlim([0, 3])
ax1.tick_params(labelsize=12)
ax1.set_xlabel(r'$x$',fontsize=15)
ax1.set_ylabel(r'$f_{1}$',fontsize = 15)
ax1.plot(sample_x, q1r1f1.flatten(), color='mediumblue')

ax2 = fig1.add_subplot(212)
ax2.set_xlim([0, 3])
ax2.tick_params(labelsize=12)
ax2.set_xlabel(r'x',fontsize=15)
ax2.set_ylabel(r'$f_{2}$',fontsize = 15)
ax2.plot(sample_x, q1r1f2.flatten(), color='mediumblue')

fig1.savefig('q1r1.png')
```
